Interpolasi bertujuan untuk mencari nilai suatu variabel yang hilang pada rentang data yang diketahui. Metode yang digunakan adalah metode lagrange yang ditemukan Joseph Louis Lagrange.
Kita akan mengeksplorasi bagaimana menerapkan metode interpolasi ke serangkaian nilai x dan y, serta bagaimana memplot kurva yang dihasilkan dari polinomial lagrange dengan titik data yang diberikan. 

Algoritma:
1. Mulai
2. Tentukan data pada nilai x dan nilai y
3. Tentukan titik interpolasi x yang akan dicari nilai y-nya
4. Buat sebuah array kosong untuk menampung hasil perhitungan titik interpolasi y
5. Lakukan perhitungan dengan menggunakan rumus Lagrange secara berulang

![](https://latex.codecogs.com/svg.image?P_n(x)&space;=&space;\sum_{i=0}^{n}&space;y_i&space;\cdot&space;L_i(x))
6. Lakukan langkah (5) sebanyak n data
7. Masukkan setiap hasil pehitungan langkah (5) ke array yang kosong di langkah ke (4)
8. Selesai

# **HARD CODE**

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([5, 10, 15, 20, 25, 30, 35, 40], float)
y = np.array([40, 30, 25, 40, 18, 20, 22, 15], float)
x_in = float(input("Masukkan titik interpolasi: "))

xplt = np.linspace(x[0],x[-1])
yplt = np.array([], float)

for i in range(len(x)):
    y_out = 0
    for xi, yi in zip(x,y):
        y_out += yi * np.prod((x_in - x[x != xi])/(xi - x[x != xi]))
print("Untuk X %.3f, Y = %.3f" % (x_in, y_out))

for xp in xplt: # perulangan setiap data x di sepanjang xplt
    yp = 0 # tampungan

    for xi, yi in zip(x,y):
        print(f"{y_out:.2f}\t=\t{y_out:.2f} * {x_in}-{x[x != xi]}/{xi}-{x[x != xi]}\n")
        yp += yi * np.prod((xp - x[x != xi])/(xi - x[x != xi]))
    yplt = np.append(yplt,yp)

print("Interpolasi")
for m, n in zip(xplt, yplt):
    print("Untuk X %.3f, Y = %.3f" % (m, n))

# plot
plt.plot(x, y, "o", xplt,yplt,'b-')
plt.plot(x_in, y_out, "ro")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
